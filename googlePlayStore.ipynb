```python
contentList = artsdf["Content Rating"].tolist()
beautyList = beautydf["Content Rating"].tolist()
carList = carsdf["Content Rating"].tolist()
bookList=bookdf["Content Rating"].tolist()
workList=workdf["Content Rating"].tolist()
comicList=comicdf["Content Rating"].tolist()
commList=commdf["Content Rating"].tolist()
datingList=datingdf["Content Rating"].tolist()
eduList=edudf["Content Rating"].tolist()
dic = {"Everyone":"red", "Teen":"green", "Everyone 10+":"blue","Mature 17+":"purple","Adults only 18+":"yellow"}
colorList = [dic.get(colors, colors) for colors in contentList]
comicColour=[dic.get(colors, colors) for colors in comicList]
colourList=[dic.get(colors,colors) for colors in beautyList]
carColourList=[dic.get(colors,colors) for colors in carList]
bookColour =[dic.get(colors,colors) for colors in bookList]
workColour =[dic.get(colors,colors) for colors in workList]
commColour=[dic.get(colors,colors) for colors in commList]
datingColour=[dic.get(colors,colors) for colors in datingList]
eduColour=[dic.get(colors,colors) for colors in eduList]
```

```python
fig, ax = plt.subplots() 
ax.scatter(artsdf["Rating"],artsdf["Reviews"],c=colorList, s =30,alpha=0.1)
ax.set_title('Reviews vs. Rating for apps in category: Art and Design', loc='right', fontsize=14, fontweight = "bold")
ax.set_xlabel("Rating")
ax.set_ylabel("Reviews")
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
```

```python
fig, ax = plt.subplots() 
ax.scatter(beautydf["Rating"],beautydf["Reviews"],c=colourList, s =30,alpha=0.3)
ax.set_title('Reviews vs. Rating for apps in category: beauty', loc='right', fontsize=14, fontweight = "bold")
ax.set_xlabel("Rating")
ax.set_ylabel("Reviews")
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.set_ylim(0,60000)
```

```python
fig, ax = plt.subplots() 
ax.scatter(carsdf["Rating"],carsdf["Reviews"],c=carColourList, s =30,alpha=0.2)
ax.set_title('Reviews vs. Rating for apps in category: cars', loc='right', fontsize=14, fontweight = "bold")
ax.set_xlabel("Rating")
ax.set_ylabel("Reviews")
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
```

```python
fig, ax = plt.subplots(nrows = 3, ncols = 3, sharex = True,sharey=True,figsize = (12,7)) 

ax[0][0].scatter(beautydf["Rating"],beautydf["Reviews"],c=colourList, s =10,alpha=0.1)
ax[0][1].scatter(artsdf["Rating"],artsdf["Reviews"],c=colorList, s =10,alpha=0.1)
ax[0][2].scatter(carsdf["Rating"],carsdf["Reviews"],c=carColourList, s =10,alpha=0.1)
ax[1][0].scatter(bookdf["Rating"],bookdf["Reviews"],c=bookColour, s =10,alpha=0.1)
ax[1][1].scatter(workdf["Rating"],workdf["Reviews"],c=workColour, s =10,alpha=0.1)
ax[1][2].scatter(comicdf["Rating"],comicdf["Reviews"],c=comicColour,s=10,alpha=0.1)
ax[2][0].scatter(commdf["Rating"],commdf["Reviews"],c=commColour,s=10,alpha=0.1)
ax[2][1].scatter(datingdf["Rating"],datingdf["Reviews"],c=datingColour,s=10,alpha=0.1)
ax[2][2].scatter(edudf["Rating"],edudf["Reviews"],c=eduColour,s=10,alpha=0.1)
fig.suptitle("Reviews vs. Rating for apps in different category", fontsize = 14, fontweight = "bold")
titles = ["Beauty","Art and Design","Auto and Vehicles","Books and Reference","Business","Comics","Commiction","Dating",'Education']
count=0
for var in ax:
    for i in var:
        i.spines["right"].set_visible(False) 
        i.spines["top"].set_visible(False)
        i.set_ylabel("Reviews")
        i.set_title(titles[count],fontsize=10)
        count=count+1

ax[2][0].set_xlabel("Ratings")
ax[2][1].set_xlabel("Ratings")
ax[2][2].set_xlabel("Ratings")



plt.show()
```
